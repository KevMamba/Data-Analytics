```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pylab import *
import os
```

```python
shampoo = pd.read_csv("C:/Users/kevin/Desktop/PES University/Junior Year/DA/Shampoo.csv")
df = pd.DataFrame(shampoo)
#print ((df['Sales of shampoo over a three year period']).to_string(index=False))
```

```python
#Moving average in intervals of 3
short_rolling = df['Sales of shampoo over a three year period'].rolling(window=3).mean()
short_rolling.fillna(0, inplace=True)
plt.plot(short_rolling)
plt.xticks(range(len(df['Month'])), df['Month'])
plt.title('Moving average (in intervals of 3)')
plt.ylabel('Shampoo Sales')
plt.xlabel('Time (in months)') 
plt.show()
```

```python
#Moving average in intervals of 5
short_rolling = df['Sales of shampoo over a three year period'].rolling(window=5).mean()
short_rolling.fillna(0, inplace=True)
plt.plot(short_rolling)
plt.xticks(range(len(df['Month'])), df['Month'])
plt.title('Moving average (in intervals of 5)')
plt.ylabel('Shampoo Sales')
plt.xlabel('Time (in months)') 
plt.show()
```
